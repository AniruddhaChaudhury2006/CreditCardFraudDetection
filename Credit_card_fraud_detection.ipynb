```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.tail() #displaying last 5 rows
```

```python
credit_card_data.info() #dataset information
```

```python
credit_card_data.isnull().sum() #checking for null values
```

```python
credit_card_data['Class'].value_counts() #distribution of legit transactions & fraudulent transactions
```

**HIGHLY UNBALANCED DATASET WHERE 0 DENOTES NORMAL TRANSACTION AND 1 DENOTES FRAUD ONE**

```python
# separating data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe() #statistical measures of data
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()   #comparing values for both transactions
```

```python
legit_sample = legit.sample(n=415) # building sample dataset containing similar distribution of normal and fraudulent transactions (under sampling)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis = 1) #SPLITTING DATASET INTO FEATURES AND TARGETS
Y = new_dataset['Class']
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2) #splitting dataset into training and testing data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression() #model training using logistic regression
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train) #evaluating model using accuracy score
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on training data : ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy on test data : ", test_data_accuracy)
```
